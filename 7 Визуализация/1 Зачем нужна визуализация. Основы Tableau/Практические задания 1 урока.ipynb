## 2. Bar chart хорошо решает следующую задачу:

* Узнать точное число для точки
* **Сравнить несколько категорий между собой**
* Понять кластеры данных
* Увидеть среднее значение


## 3. Scatter plot хорошо решает задачу/задачи:

* **Проанализировать распределение данных по двум переменным**
* **Найти кластеры по двум переменным**
* **Увидеть выбросы в данных**
* **Понять есть ли корреляция между двумя переменными**


## 4. Таблица хорошо решает задачу/задачи:

* Понять есть ли корреляция между двумя переменными
* Увидеть выбросы в данных
* **Сравнить несколько разрезов по разным метрикам**
* **Узнать точное значение данных**
