## 5. Функция `ATTR()`:

* Вычисляет атрибуты переданных параметров и выводит их в качестве текста. Если не может рассчитать атрибут, то возвращает знак звёздочки.
* **Возвращает значение выражения, если оно имеет одно значение для всех строк. В противном случае возвращает знак звёздочки.** 
<br>_**(для всех строк внутри группы)**_
* Вычисляет среднее значение, при этом игнорируются нуль и отрицательные значения. Если все значения нули или отрицательные, то возвращает знак звёздочки.
* Возвращает количество элементов в группе. Нулевые значения не учитываются, если все значения нули, то возвращает знак звёздочки.

## 6. Какая формула посчитает процентный прирост метрики (Metric — расчетное поле с агрегацией)?

* [Metric] / TOTAL([Metric])
* ZN([Metric]) / ABS(LOOKUP(ZN([Metric]), -1))
* **(ZN([Metric]) - LOOKUP(ZN([Metric]), -1)) / ABS(LOOKUP(ZN([Metric]), -1))**
* ZN([Metric]) - LOOKUP(ZN([Metric]), -1)

## 7. Какая формула позволит динамически переключать гранулярность даты Date? ([Scale] — параметр со значениями "day", "week", "month")

* DATETRUNC("month",[Date])
* DATETRUNC([Date],[Scale])
* DATEPART([Date],[Scale])
* **DATETRUNC([Scale],[Date])**
