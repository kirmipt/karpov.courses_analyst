Рассчитайте p-уровень значимости для проверки нулевой гипотезы о том, что среднее значение баллов в 2015 году равняется 115.**

* p < 0.05 – принимаем нулевую гипотезу
* p > 0.05 – принимаем нулевую гипотезу
* p > 0.05 – отклоняем нулевую гипотезу
* **p < 0.05 – отклоняем нулевую гипотезу**
