**Задание 1**

**Два человека одновременно стреляют по мишени. Вероятность попадания по мишени у первого стрелка равна 0,5, у второго — 0,8. Какова вероятность того, что в мишени будет только одна пробоина?**

Примечание: в этой и последующих задачах десятичным разделителем должна быть точка. 

(1попал и 2не попал) или (2попал и 1не попал):

**Ответ:** 0.5 * (1 - 0.8) + 0.8 * (1 - 0.5) = `0.5`

<br>_кстати все случаи в сумме 1
<br>0.5*0.8 + 0.5*0.2 + 0.5*0.8 + 0.5*0.2 = 1_

<br>_можно обобщить этот факт на n незав. событий по индукции_

<br>_правда для n=1: a + 1 - a = 1_

<br>_из версии для n следует для n+1:
<br>[a1 или (1-a1)] * .. * [an или (1-an)]  +  похожий член  +... <br>таких членов столько, сколько возможностей выбрать ai или 1-ai
<br>добавилось * [a(n+1) или (1-a(n+1))] в каждое перемножение и перемножений стало в 2 раза больше:_

<br>_sum.for.n+1 = sum.for.n * a(n+1) + sum.for.n * (1-a(n+1)) = 
<br>= sum.for.n * [a(n+1) + 1 - a(n+1))] = sum.for.n = 1_
